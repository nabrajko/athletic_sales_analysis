```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
file = './Resources/athletic_sales_2020.csv'
sales2020_df = pd.read_csv(file)
file1 = './Resources/athletic_sales_2021.csv'
sales2021_df = pd.read_csv(file1)

```

```python
# Display the 2020 sales DataFrame
sales2020_df.head()
```

```python
# Display the 2021 sales DataFrame
sales2021_df.head()
```

#### Check the data types of each DataFrame

```python
sales2020_df.dtypes
```

```python
# Check the 2020 sales data types.
sales2020_df.info(verbose=True)
```

```python
# Check the 2021 sales data types.
sales2021_df.info(verbose=True)
```

#### Combine the sales data by rows.

```python
combined_df = pd.concat([sales2020_df, sales2021_df], axis=0).reset_index( drop=True)
combined_df.head()
```

```python
# Check if any values are null
combined_df.isna().sum()
```

```python
# Check the data type of each column
combined_df.info(verbose=True)
```

```python
# Convert the "invoice_date" to a datetime 
combined_df['invoice_date']=pd.to_datetime(combined_df['invoice_date'], errors='coerce')
```

```python
# Confirm that the "invoice_date" data type has been changed.
combined_df.info()
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
Total_Products_Sold=combined_df.groupby(['region','state','city'])['units_sold'].agg(Total_Products_Sold =('sum'))


# Show the top 5 results.
Total_Product_Sorted=Total_Products_Sold.sort_values(by='Total_Products_Sold' , ascending=False)
Total_Product_Sorted.head()
```

```python
combined_df.info()
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.

combined_df_pivot = pd.pivot_table(combined_df, values="units_sold",index=["region","state","city"] ,aggfunc='sum' )
combined_df_pivot.head()

# Rename the "units_sold" column to "Total_Products_Sold"
Total_Product_pivot=combined_df_pivot.rename(columns={"units_sold":"Total_Products_Sold"})

# Show the top 5 results.
Total_Product_pivot.sort_values(by='Total_Products_Sold' , ascending=False).head()
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
Total_product_sales=combined_df.groupby(by=["region","state","city"])['total_sales'].agg(Total_Sales='sum')

# Rename the "total_sales" column to "Total Sales"

# Show the top 5 results.
Total_product_sales.sort_values(by='Total_Sales', ascending=False).head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.

Total_Sales_Pivot=pd.pivot_table(combined_df, values="total_sales",index=["region","state","city"] ,aggfunc='sum' )

# Optional: Rename the "total_sales" column to "Total Sales"
Total_Sales_rename=Total_Sales_Pivot.rename(columns={"total_sales":"Total_Sales"})

# Show the top 5 results.
Total_Sales_rename.sort_values(by='Total_Sales' , ascending=False).head()
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
# Rename the "total_sales" column to "Total Sales"
Total_product_sold_retailer=combined_df.groupby(by=["retailer","region","state","city"])['total_sales'].agg(Total_Sales='sum')

# Show the top 5 results.
Total_product_sold_retailer.sort_values(by='Total_Sales', ascending=False).head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.

Total_Sales_Pivot_retailer=pd.pivot_table(combined_df, values="total_sales",index=["retailer","region","state","city"] ,aggfunc='sum' )

# Optional: Rename the "total_sales" column to "Total Sales"
Total_Sales_Pivot_retailer=Total_Sales_Pivot_retailer.rename(columns={"total_sales":"Total_Sales"})

# Show the top 5 results.
Total_Sales_Pivot_retailer.sort_values(by='Total_Sales', ascending=False).head()
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
combined_df.head()
women_athletic_footwear=combined_df.loc[combined_df["product"]=="Women's Athletic Footwear"]
women_athletic_footwear.head()
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"

Total_women_athletic_footwear=women_athletic_footwear.groupby(by=["retailer","region","state","city"])['units_sold'].agg(Womens_Footwear_Units_Sold='sum')

# Show the top 5 results.

Total_women_athletic_footwear.sort_values(by='Womens_Footwear_Units_Sold', ascending=False).head()
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.

Total_women_athletic_footwear_pivot=pd.pivot_table(women_athletic_footwear, values="units_sold",index=["retailer","region","state","city"] ,aggfunc='sum' )

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
Total_women_athletic_footwear_pivot=Total_women_athletic_footwear_pivot.rename(columns={"units_sold":"Womens_Footwear_Units_Sold"})

# Show the top 5 results.
Total_women_athletic_footwear_pivot.sort_values(by="Womens_Footwear_Units_Sold" , ascending=False).head()
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
Total_women_athletic_footwear=pd.pivot_table(women_athletic_footwear, values="total_sales",index=["invoice_date"] ,aggfunc='sum' )
Total_women_athletic_footwear.head(10)

# Optional: Rename the "total_sales" column to "Total Sales"


# Show the table.
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
daily_bin=Total_women_athletic_footwear.resample('D').mean()

# Sort the resampled pivot table in descending order on "Total Sales".
daily_bin_bin_sorted=daily_bin.sort_values(by="total_sales", ascending=False)
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
weekly_bin=Total_women_athletic_footwear.resample('W').mean()

# Sort the resampled pivot table in descending order on "Total Sales".
weekly_bin_bin_sorted=weekly_bin.sort_values(by="total_sales", ascending=False)
```
